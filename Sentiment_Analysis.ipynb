## A. Crawling Data

```python
# Read the CSV file into a DataFrame
# The index_col=0 parameter specifies that the first column in the CSV file will be used as the row index
df = pd.read_csv('/content/data_real.csv', index_col=0)

# Display the first 5 rows of the DataFrame to get a quick overview of the data
df.head()
```

```python
# Create a copy of the DataFrame 'df'
# This ensures that 'df_copy' is an independent copy of 'df'
# Changes made to 'df_copy' will not affect 'df', and vice versa
df_copy = df.copy()
```

## Labeling

```python
# Initialize an empty list to store the labels
label = []

# Iterate over each row in the DataFrame 'df_copy'
for index, row in df_copy.iterrows():
    # Check the value of the 'Rating' column for each row
    if row['Rating'] == 1 or row['Rating'] == 2:
        # If the rating is 1 or 2, assign label 0
        label.append(0)
    else:
        # For other ratings, assign label 1
        label.append(1)

# Add the 'label' list as a new column in the DataFrame 'df_copy'
df_copy["label"] = label
```

```python
df_copy.head()
```

## C. Preprocessing Text

```python
# Convert string in Ulasan column to lowercase
df_copy['Ulasan'] = df_copy['Ulasan'].str.lower()

df_copy.head()
```

## C.1. Normalization

```python
norm = { " dgn " : " dengan ", " gue ": " saya ", " dgn ":" dengan ", "bgmn ":" bagaimana ", ' tdk':' tidak ', ' blum ':' belum ', 'mantaaaaaaaappp':' bagus ', ' josss ':' bagus ', ' thanks ': ' terima kasih ', 'fast':' cepat ', ' dg ':' dengan ', 'trims':' terima kasih ', 'brg':' barang ', 'gx':' tidak ', ' dgn ':' dengan ', ' recommended':' rekomen ', 'recomend':' rekomen ', 'good':' bagus ' }
def normalisasi(str_text):
  for i in norm:
    str_text = str_text.replace(i, norm[i])
  return str_text

df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: normalisasi(x))
```

```python
df_copy.head()
```

## C.2. Stopwords

```python
# Corrected import statements for the Sastrawi library
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary

more_stop_words = []

# Initialize stop word remover with default stop words
stop_words = StopWordRemoverFactory().get_stop_words()
new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

# Function to remove stop words from a given text
def stopword(str_text):
    # Remove stop words from the text
    str_text = stop_words_remover_new.remove(str_text)
    return str_text

# Apply the 'stopword' function to the 'Ulasan' column of the DataFrame
# This will remove stop words from each entry in the 'Ulasan' column
df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: stopword(x))
```

```python
df_copy.head()
```

## C.3. Tokenization

```python
# Apply a lambda function to each entry in the 'Ulasan' column
# The lambda function splits each text entry into a list of words (tokenization)
tokenized = df_copy['Ulasan'].apply(lambda x: x.split())

# Display the result
# 'tokenized' now contains a DataFrame where each entry in the 'Ulasan' column has been split into a list of words
tokenized
```

## C.4. Stemming

```python
# Import the StemmerFactory class from Sastrawi library
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Define a function to perform stemming on a list of words
def stemming(Ulasan):
    # Create an instance of StemmerFactory
    factory = StemmerFactory()
    # Create a stemmer object using the factory
    stemmer = factory.create_stemmer()
    # Initialize an empty list to store the stemmed words
    do = []
    # Iterate over each word in the input list 'Ulasan'
    for w in Ulasan:
        # Perform stemming on the word
        dt = stemmer.stem(w)
        # Append the stemmed word to the list 'do'
        do.append(dt)
    # Join the list of stemmed words into a single string with spaces between words
    d_clean = " ".join(do)
    # Print the cleaned and stemmed text
    print(d_clean)
    # Return the cleaned and stemmed text
    return d_clean

# Apply the 'stemming' function to each list of words in the 'tokenized' DataFrame
# This will stem each word and join them back into a single string
tokenized = tokenized.apply(stemming)

# Save the cleaned data to a CSV file named 'dataclean.csv'
# 'index=False' ensures that the index is not written to the CSV file
tokenized.to_csv('dataclean.csv', index=False)

# Read the cleaned data from the CSV file
# 'encoding="latin1"' is used to handle encoding issues if any
data_clean = pd.read_csv('dataclean.csv', encoding='latin1')

# Display the first 5 rows of the cleaned data to verify the results
data_clean.head()
```

## D. Gather/Concatenate 2 attributes

```python
# Read the CSV file '/content/dataclean.csv' into a DataFrame 'at1'
# This DataFrame contains the cleaned text data
at1 = pd.read_csv('/content/dataclean.csv')

# Read the CSV file '/content/data_real.csv' into a DataFrame 'at2'
# This DataFrame contains the original data with labels
at2 = pd.read_csv('/content/data_real.csv')

# Extract the 'label' column from DataFrame 'at2'
# This column contains labels or categories associated with the data
att2 = at2['label']

# Concatenate DataFrames 'at1' and 'att2' along the columns (axis=1)
# This combines 'at1' (cleaned data) with 'att2' (labels) into a single DataFrame
# 'axis=1' specifies that the concatenation is done along the columns (horizontally)
# The resulting DataFrame 'result' will have the cleaned text data from 'at1'
# combined with the label column from 'att2', aligning them side by side based on their row index.
result = pd.concat([at1, att2], axis=1)
```

```python
result.head()
```

## E. Word Calculation using TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```python
# Extract the 'Ulasan' column from the 'result' DataFrame
# This column contains the cleaned review text
Ulasan = result['Ulasan']
```

```python
# Check how many missing values (NaN) are in the 'Ulasan' column
# This helps ensure there are no missing values before proceeding
Ulasan.isnull().sum()
```

```python
# Fill missing values with the string 'tidak ada komentar'
# This replaces any NaN values with a placeholder text
Ulasan = Ulasan.fillna('tidak ada komentar')
```

```python
# Create a CountVectorizer object
# CountVectorizer is used to convert text into a matrix of token counts
cv = CountVectorizer()

# Fit the CountVectorizer to the 'Ulasan' data and transform it into a matrix of token counts
# This builds a vocabulary from the data and converts text into numbers
term_fit = cv.fit(Ulasan)

# Print the number of unique words found
# This shows how many different words are in the vocabulary
print(len(term_fit.vocabulary_))
```

```python
# Print the vocabulary dictionary
# This shows how each word is mapped to a unique index
term_fit.vocabulary_
```

```python
# Column 1 = the number of docs
# Column 2 = the location of the word
# Column 3 = TF result

# Transform all reviews into a matrix of token counts
# This creates a matrix where each row is a review and each column is a word from the vocabulary
term_frequency_all = term_fit.transform(Ulasan)

# Print the token count matrix for all reviews
# This shows the numeric representation of the text reviews
print(term_frequency_all)
```

```python
# Select the second review (index 1) for analysis
# This is a specific review text that we will analyze separately
ulasan_tf = Ulasan[1]

# Print the selected review text
# This lets us see the text that will be converted into numbers
print(ulasan_tf)
```

```python
# Transform the selected review into a matrix of token counts
# This converts the specific review text into the same format as the other reviews
term_frequency = term_fit.transform([ulasan_tf])

# Print the token count matrix for the selected review
# This shows how the selected review is represented numerically
print(term_frequency)
```

```python
dokumen = term_fit.transform(Ulasan)
tfidf_transformer = TfidfTransformer().fit(dokumen)
print(tfidf_transformer.idf_)

tfidf = tfidf_transformer.transform(term_frequency)
print(tfidf)
```
